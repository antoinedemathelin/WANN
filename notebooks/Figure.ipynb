```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(0, 2, 6)
ys = np.concatenate((np.array([1, 1, 1]), xs[3:]))
xt = np.linspace(1, 3, 6)
   
   
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (16, 6))

ax1.fill_between([0, 2], [-0.5, -0.5], [-1, -1], alpha=0.5, label=r"$Q$")
ax1.fill_between([1, 3], [-0.5, -0.5], [-1, -1], alpha=0.5, label=r"$P$")
ax1.plot(xs, xs, "o", ms=15, alpha=0.9, label=r"$(x_i, y_i) \in \mathcal{S}$")
ax1.plot(xt, xt, "s", ms=15, alpha=0.9, label=r"$(x_i', y_i') \in \mathcal{T}$")
ax1.plot([-0.5, 3.5], [-0.5, 3.5], ls="--", c="black", alpha=0.9, label=r"predictions")
ax1.set_xlim(-0.5, 3.5)
ax1.set_ylim(-1, 3.5)
ax1.set_yticks([0., 1., 2., 3.])
ax1.set_xticks([0., 1., 2., 3.])
ax1.legend(fontsize=18, loc="upper left")
for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(16)
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(16)
ax1.set_title("Before features transformation", fontsize=22)
   
   
ax2.fill_between([0, 2], [-0.5, -0.5], [-1, -1], alpha=0.5, label=r"$Q$")
ax2.fill_between([0, 2], [-0.5, -0.5], [-1, -1], alpha=0.5, label=r"$\phi(P)$")
ax2.plot(xs, xs, "o", ms=15, alpha=0.9, label=r"$(x_i, y_i)$")
ax2.plot(xt-1, xt, "s", ms=15, alpha=0.9, label=r"$(\phi(x_i'), y_i')$")
ax2.plot([-0.5, 3.5], [0., 4.], ls="--", c="black", alpha=0.9, label=r"predictions")
ax2.set_xlim(-0.5, 3.5)
ax2.set_ylim(-1, 3.5)
ax2.set_yticks([0., 1., 2., 3.])
ax2.set_xticks([0., 1., 2., 3.])
ax2.legend(fontsize=18)
for tick in ax2.xaxis.get_major_ticks():
    tick.label.set_fontsize(16)
for tick in ax2.yaxis.get_major_ticks():
    tick.label.set_fontsize(16)
ax2.set_title("After features transformation", fontsize=22)


plt.subplots_adjust(wspace=0.1)
plt.show()
```
